# Marketing ROI Analysis - Simple Linear Regression

## Project Overview
This project analyzes a marketing dataset to determine which advertising channel (TV, Radio, or Social Media) generates the highest return on investment (ROI) using Simple Linear Regression.

## Business Question
**Which marketing channel should we invest more in to maximize sales?**

## Dataset
- **Source**: Marketing and sales data
- **Rows**: 4,546 (after cleaning)
- **Columns**: TV, Radio, Social_Media, Sales

## Key Findings

### Best Performing Channel: TV (for sales volume)
- **R-squared**: 99.90% of sales explained
- **Coefficient**: $3.56 sales per $1 spent
- **P-value**: Statistically significant (p < 0.05)

### Highest ROI Channel: Social Media
- **ROI**: $22.19 per $1 spent
- **R-squared**: 27.82%
- **P-value**: Statistically significant (p < 0.05)

### Model Equation
**Sales = -0.13 + 3.5615 × TV**

## Results Summary

| Channel | ROI per $1 | R-squared | Significant? |
|---------|-----------|-----------|--------------|
| TV | $3.56 | 99.90% | ✅ Yes |
| Radio | $8.36 | 75.45% | ✅ Yes |
| Social Media | $22.19 | 27.82% | ✅ Yes |

## Business Recommendation

### Hybrid Strategy:
1. **MAINTAIN TV BUDGET** - Drives the bulk of sales (99.9% explanatory power)
2. **INCREASE SOCIAL MEDIA** (20-30%) - Highest ROI ($22.19 per $1)
3. **OPTIMIZE RADIO** - Solid ROI ($8.36 per $1)

### Example Budget Allocation ($10,000):
- TV (45%): $4,500
- Social Media (30%): $3,000
- Radio (25%): $2,500
- **Projected Return**: $92,310
